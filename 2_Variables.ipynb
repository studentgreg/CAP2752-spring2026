```python
a = 10
b = 3

print(a+b) # addition
print(a*b) # multiplication
print(a-b) # subtraction
print(a/b) # division, always results in a float  10/2 = 5.0
# the reason is so that the type of ALL results are the same
print(a//b) # floor division
print(a**b) # a to the power of b
print(a%b) # modulo or the remainder of the division of a by b
```

```python
x = 10
x = x + 2
print(x)

x += 3 # this is the same as writing x = x + 3
print(x)

y = 9.2
y -= 3 # this is the same as writing y = y - 3
print(y)
```

### Boolean and Comparison Operators

-< less than

-> greater than

-== equal to

-<= less or equal

->= greater or equal

-!= not equal

-True

-False

-or

-and

```python
a = 7 > 8
print(a)

b = 10 > (3*3)
print(b)

c = (7 > 8) or (9 > 8)
print(c)

d = len("greg") >= 4
print(d)
```

```python
university = "fiu "

print(type(university)) # returns the type which is string
print(len(university)) # returns how many characters which is 4
print(university.title()) # first letter upper case
print(university.upper()) # all letters upper case
print(university.lower()) # all letters lower case
print(university[0]) # accessing index 0, which is the first character
print(university[1]) # second character
print(university[2]) # third character
print(university[3]) # fourth character
```
