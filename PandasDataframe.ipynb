```python
# Pandas dataframe (for tabular data such as a csv file)
import pandas as pd
from pprint import pprint # optional print method (pretty print)

personal_info = {
    "Name":["Greg","Alice","Leo","Rebeca","David"],
    "Age":[23,35,45,25,30],
    "Height (cm)":[165,175,160,180,170],
    "Weight (kg)":[65,85,70,90,60],
}
print(type(personal_info))
pprint(personal_info)
```

```python
# Converting python dictionary to pandas datrame

df = pd.DataFrame(personal_info)
df # no need to use print because Google Colab prints the table better
```

```python
df["Age"]
```

```python
df.columns
```

```python
df.shape # rows and columns
```

```python
print("Average height", df["Height (cm)"].mean())

print("Minimum height", df["Height (cm)"].min())

print("Maximum height", df["Height (cm)"].max())

print("Variance of height", df["Height (cm)"].var())

print("Standard deviation of height", df["Height (cm)"].std())
```

```python
# 1. Save the maximum value of weight in a variable
maxWeight = df["Weight (kg)"].max()
print(maxWeight)

# 2. Save the minimum value of weight in another variable
minWeight = df["Weight (kg)"].min()
print(minWeight)

# 3. Calculate the weight range
rangeWeight = maxWeight - minWeight
print(rangeWeight)
```

```python
# Create a new column for the height in meters

df["Height (m)"]=df["Height (cm)"]/100
df
```

```python
# 4. Add a new column for the weight in lbs
df["Weight (lbs)"]=df["Weight (kg)"]*2.205
df

# 5. Add a new column for the BMI (body mass index)
df["BMI"]=df["Weight (kg)"]/(df['Height (m)']**2)
df
```

```python
#Filtering my table

condition1 = df["Age"]>30
df[condition1]
```

```python
condition2 = df["Weight (kg)"]<=70
df[condition2]
```

```python
df[condition1 & condition2] # both conditions 1 and 2
```

```python
# Condition1 (age > 30) and Weight (kg) > 70
df[condition1 & ~condition2]
```

```python
# Alternatively, you can create a new condition
condition3 = df["Weight (kg)"]>70
df[condition1 & condition3]
```
